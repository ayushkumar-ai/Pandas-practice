```python
import pandas as pd

data = {
    "name": ["A", "B", "C", "D"],
    "age": [20, 21, None, 23],
    "marks": [85, None, 88, 90]
}

df = pd.DataFrame(data)
print(df )
print(df.isna())
print(df.isna().sum())
df_filled = df.fillna(df.mean(numeric_only=True))
print(df_filled)
df_dropped = df.dropna()
print(df_dropped)
print(df_filled.groupby("age")["marks"].mean())
```
