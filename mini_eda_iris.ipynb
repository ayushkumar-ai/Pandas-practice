```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

# Basic EDA
df.head()
df.shape
df.info()
df.describe()
df.isnull().sum()

# Analysis
df.groupby("species").mean()

# Correlation
df.corr()

# Visualization
df["sepal length (cm)"].hist()
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

## Dataset Overview

The Iris dataset is a classic dataset used for classification and exploratory data analysis.
It contains measurements of 150 iris flowers from three species:
- Iris setosa
- Iris versicolor
- Iris virginica

Each sample includes four numerical features:
- Sepal length (cm)
- Sepal width (cm)
- Petal length (cm)
- Petal width (cm)

The dataset has:
- 150 rows
- 5 columns (including species)
- No missing values

## Key Observations

- Sepal length values range approximately from 4.3 cm to 7.9 cm.
- The distribution of sepal length shows multiple peaks, indicating differences among species.
- Most samples have sepal lengths between 5.0 cm and 6.5 cm.
- There are fewer samples with very small or very large sepal lengths.
- The variation in sepal length suggests it can help distinguish between different iris species.

## Conclusion

This exploratory data analysis shows that the Iris dataset is clean and well-structured.
Basic statistical analysis and visualization reveal meaningful patterns in the data.
Sepal length demonstrates noticeable variation across samples and can be a useful feature
for classification tasks.
