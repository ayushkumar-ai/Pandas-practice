```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame
df['species'] = iris.target


df.head()
df.shape
df.info()

plt.plot(df['sepal length (cm)'])
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()

plt.hist(df['petal length (cm)'], bins=20)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()

plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()

mean_petal = df.groupby('species')['petal length (cm)'].mean()

plt.bar(mean_petal.index, mean_petal.values)
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].hist(df['sepal width (cm)'])
ax[0].set_title("Sepal Width Distribution")

ax[1].hist(df['petal width (cm)'])
ax[1].set_title("Petal Width Distribution")

plt.show()

```


Key Observations

1. Petal features show clearer variation compared to sepal features.
2. Scatter plot shows a positive relationship between sepal and petal length.
3. Different species have noticeably different average petal lengths.


Conclusion

feature distributions and relationships. The visual patterns suggest
